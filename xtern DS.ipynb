```python
import requests
import pandas as pd
import json
import numpy as np
```

```python
### Restaurants in Indianapolis
```

```python
API_KEY = %env API_KEY
print(API_KEY)
url = "https://maps.googleapis.com/maps/api/place/textsearch/json?query=restaurants%20in%20Indianapolis&key=AIzaSyB5NDXHCfd8RcHoVLCUpPtu3A3HXnH9Iok"
response = requests.get(url, headers={}, data={})
```

```python
json_data = json.loads(response.content)["results"]
```

```python
restaurant_name = []
restaurant_website = []
restaurant_address = []
restaurant_rating = []
restaurant_reviewNum = []
restaurant_types = []

for i in json_data: #referenced https://towardsdatascience.com/how-to-use-the-google-places-api-for-location-analysis-and-more-17e48f8f25b1
    details_url = "https://maps.googleapis.com/maps/api/place/details/output?place_id=" + i["place_id"] + "&key=" + API_KEY
    try:
        details_response = requests.get(url, headers={}, data={})
        print(details_response.status)
        restaurant_website.append(details_response.content.results["website"])
        restaurant_reviewNum.append(details_response.content.results["user_ratings_total"])
    except:
        restaurant_website.append(np.nan)
        restaurant_reviewNum.append(np.nan)
    try:
        restaurant_name.append(i["name"])
    except:
        restaurant_name.append(np.nan)
    try:
        restaurant_address.append(i["formatted_address"])
    except:
        restaurant_address.append(np.nan)
    try:
        restaurant_rating.append(i["rating"])
    except:
        restaurant_rating.append(np.nan)
    try:
        restaurant_types.append(i["types"])
    except:
        restaurant_types.append(np.nan)
```

```python
restaurant_dict = {"r_name": restaurant_name, "r_address": restaurant_address,
                   "r_rating": restaurant_rating, "r_review_num": restaurant_reviewNum,
                   "r_website": restaurant_website,"r_types": restaurant_types}
restaurant_df = pd.DataFrame(restaurant_dict)
print(restaurant_df)
```
